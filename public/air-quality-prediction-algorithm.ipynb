```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.metrics import mean_squared_error, silhouette_score
from sklearn.cluster import KMeans
from scipy.signal import find_peaks

# Load the dataset
file_path = 'Downloads/Raw_data_1Hr_2023_site_288_Velachery_Res._Area_Chennai_CPCB_1Hr (1).csv'
data = pd.read_csv(file_path)

# Convert the Timestamp column to datetime format and set as index
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data = data.set_index('Timestamp')

# Exploratory Data Analysis (EDA)
def eda(data):
    print("Basic Information:")
    print(data.info())
    print("\nStatistical Summary:")
    print(data.describe())
    
    print("\nMissing Values:")
    print(data.isna().sum())
    
    # Plotting time series data
    plt.figure(figsize=(14, 7))
    sns.lineplot(data=data[['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)']])
    plt.title('Time Series Plot of PM2.5, PM10, and NO2')
    plt.xlabel('Timestamp')
    plt.ylabel('Concentration (µg/m³)')
    plt.legend(['PM2.5', 'PM10', 'NO2'])
    plt.show()
    
    # Pairplot for selected features
    sns.pairplot(data[['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)', 'RH (%)', 'WS (m/s)', 'WD (deg)']])
    plt.show()
    
    # Correlation matrix
    plt.figure(figsize=(14, 12))
    corr_matrix = data.corr()
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
    plt.title('Correlation Matrix')
    plt.show()

eda(data)

# Preprocessing
# Impute missing values with mean for numerical columns
data.fillna(data.mean(), inplace=True)

# Drop columns with no records if necessary
data.dropna(axis=1, how='all', inplace=True)

# Reduce dataset to approximately 2000 records
data = data.iloc[:2000]

# Select features and target
features = data[['RH (%)', 'WS (m/s)']]
target = data[['PM2.5 (µg/m³)']]

# Drop rows with NaN values in the target or features
data.dropna(subset=['RH (%)', 'WS (m/s)', 'WD (deg)', 'PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)'], inplace=True)

# Check for NaN or infinite values in the dataset
if np.any(np.isnan(data)) or np.any(np.isinf(data)):
    raise ValueError("Dataset contains NaN or infinite values. Please clean the data before proceeding.")

# Select features and target again to ensure cleaned data is used
features = data[['RH (%)', 'WS (m/s)', 'WD (deg)']]
target_pm25 = data['PM2.5 (µg/m³)']  # Using the 'PM2.5 (µg/m³)' column as target

# Scaling features and target independently
scaler_features = MinMaxScaler(feature_range=(0, 1))
scaler_target = MinMaxScaler(feature_range=(0, 1))
scaled_features = scaler_features.fit_transform(features)
scaled_target_pm25 = scaler_target.fit_transform(target_pm25.values.reshape(-1, 1))

# Create dataset for LSTM
def create_dataset(features, target, look_back=1):
    X, Y = [], []
    for i in range(len(features) - look_back - 1):
        a = features[i:(i + look_back), :]
        X.append(a)
        Y.append(target[i + look_back])
    return np.array(X), np.array(Y)

look_back = 10
X, Y_pm25 = create_dataset(scaled_features, scaled_target_pm25, look_back)

# Reshape input to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], X.shape[2]))

# Define test sizes
test_sizes = [0.20, 0.30]

# Cross-validation setup
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for test_size in test_sizes:
    train_rmse_list = []
    test_rmse_list = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = Y_pm25[train_index], Y_pm25[test_index]

        # Build the LSTM model
        model = Sequential([
            LSTM(50, return_sequences=True, input_shape=(look_back, X.shape[2])),
            Dropout(0.2),
            LSTM(50),
            Dropout(0.2),
            Dense(1)  # Output layer for single target variable
        ])

        # Compile the model
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model
        model.fit(X_train, y_train, epochs=50, batch_size=16, verbose=2)

        # Make predictions
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)

        # Inverse scaling for plotting and interpretation
        y_train_inv = scaler_target.inverse_transform(y_train)
        y_test_inv = scaler_target.inverse_transform(y_test)
        y_pred_train_inv = scaler_target.inverse_transform(y_pred_train)
        y_pred_test_inv = scaler_target.inverse_transform(y_pred_test)

        # Calculate RMSE for both train and test data
        train_rmse = np.sqrt(mean_squared_error(y_train_inv, y_pred_train_inv))
        test_rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_test_inv))
        train_rmse_list.append(train_rmse)
        test_rmse_list.append(test_rmse)

    # Print average RMSE over all folds
    avg_train_rmse = np.mean(train_rmse_list)
    avg_test_rmse = np.mean(test_rmse_list)
    print(f'Test size {test_size*100}% - PM2.5 Train RMSE: {avg_train_rmse:.3f}, Test RMSE: {avg_test_rmse:.3f}')

   
    # Detect and rank peaks
def detect_and_rank_peaks(data, num_categories=10):
    high_threshold = np.quantile(data, 0.90)
    low_threshold = np.quantile(data, 0.10)
    high_peaks, _ = find_peaks(data, height=high_threshold)
    low_peaks, _ = find_peaks(-data, height=-low_threshold)
    high_labels = pd.qcut(data[high_peaks], num_categories, labels=False, duplicates='drop') + 1
    low_labels = -(pd.qcut(-data[low_peaks], num_categories, labels=False, duplicates='drop') + 1)
    return high_peaks, high_labels, low_peaks, low_labels

high_peaks_train, high_labels_train, low_peaks_train, low_labels_train = detect_and_rank_peaks(y_pred_train_inv.flatten())
high_peaks_test, high_labels_test, low_peaks_test, low_labels_test = detect_and_rank_peaks(y_pred_test_inv.flatten())

plt.figure(figsize=(15, 7))

# Training Data Plot
plt.subplot(1, 2, 1)
plt.plot(range(len(y_train_inv)), y_train_inv, label='Actual')
plt.plot(range(len(y_pred_train_inv)), y_pred_train_inv, label='Predicted')
plt.scatter(high_peaks_train, y_pred_train_inv[high_peaks_train], color='red', label='High Peaks')
plt.scatter(low_peaks_train, y_pred_train_inv[low_peaks_train], color='green', label='Low Peaks')
plt.title(f'Training Data with Test Size: {test_size*100}% for PM2.5')
plt.xlabel('Time Index')
plt.ylabel('PM2.5 (µg/m³)')
plt.legend()

# Test Data Plot
plt.subplot(1, 2, 2)
plt.plot(range(len(y_test_inv)), y_test_inv, label='Actual')
plt.plot(range(len(y_pred_test_inv)), y_pred_test_inv, label='Predicted')
plt.scatter(high_peaks_test, y_pred_test_inv[high_peaks_test], color='red', label='High Peaks')
plt.scatter(low_peaks_test, y_pred_test_inv[low_peaks_test], color='green', label='Low Peaks')
plt.title(f'Test Data with Test Size: {test_size*100}% for PM2.5')
plt.xlabel('Time Index')
plt.ylabel('PM2.5 (µg/m³)')
plt.legend()

plt.show()

# Clustering with Elbow Method
def elbow_method(data):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, 11), wcss, 'bo-', markersize=8)
    plt.title('Elbow Method For Optimal Number of Clusters')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS (Within-Cluster Sum of Square)')
    plt.show()

# Running Elbow Method to find the optimal number of clusters
elbow_method(data[['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)', 'RH (%)', 'WS (m/s)', 'WD (deg)']])

# Clustering with KMeans
def clustering(data, num_clusters=4):
    # Example clustering with KMeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    data['Cluster'] = kmeans.fit_predict(data[['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)', 'RH (%)', 'WS (m/s)', 'WD (deg)']])
    
    plt.figure(figsize=(14, 7))
    sns.scatterplot(data=data, x='PM2.5 (µg/m³)', y='PM10 (µg/m³)', hue='Cluster', palette='viridis')
    plt.title('Clustering of Air Quality Data')
    plt.xlabel('PM2.5 (µg/m³)')
    plt.ylabel('PM10 (µg/m³)')
    plt.show()
    
    # Calculate silhouette score
    silhouette_avg = silhouette_score(data[['PM2.5 (µg/m³)', 'PM10 (µg/m³)', 'NO2 (µg/m³)', 'RH (%)', 'WS (m/s)', 'WD (deg)']], data['Cluster'])
    print(f'Silhouette Score: {silhouette_avg:.3f}')
    
    # Additional visualizations
    plt.figure(figsize=(14, 7))
    sns.scatterplot(data=data, x='PM2.5 (µg/m³)', y='NO2 (µg/m³)', hue='Cluster', palette='viridis')
    plt.title('Clustering of PM2.5 and NO2 with KMeans')
    plt.xlabel('PM2.5 (µg/m³)')
    plt.ylabel('NO2 (µg/m³)')
    plt.show()

    plt.figure(figsize=(14, 7))
    sns.scatterplot(data=data, x='PM2.5 (µg/m³)', y='WS (m/s)', hue='Cluster', palette='viridis')
    plt.title('Clustering of PM2.5 and Wind Speed with KMeans')
    plt.xlabel('PM2.5 (µg/m³)')
    plt.ylabel('Wind Speed (m/s)')
    plt.show()

clustering(data, num_clusters=4)
```
